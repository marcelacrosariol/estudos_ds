```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20, 6))

dataframe = pd.read_csv(
    'transfusion.data.csv',
    names=[
        'Recency (months)',
        'Frequency',
        'Monetary (c.c. blood)',
        'Time',
        'Donated in March 2007'
    ],
    skiprows=1
)

"Total: {} rows".format(len(dataframe))
```

```python
dataframe.loc[dataframe['Donated in March 2007'] == 1, 'Donated in March 2007'] = 'Yes'
dataframe.loc[dataframe['Donated in March 2007'] == 0, 'Donated in March 2007'] = 'No'
dataframe.head()
```

```python
sbn.countplot(x='Donated in March 2007', data=dataframe)
```

```python
dataframe.sort_values(by='Monetary (c.c. blood)', ascending=False)
```

```python
dataframe.sort_values(by='Monetary (c.c. blood)', ascending=True)
```
